```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/lejejefr/ECM_2526_FinalProject/refs/heads/main/Data/FullCrimeData/Crime_24-25.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(
    data=df,
    x="Longitude",
    y="Latitude",
    s=10
)

plt.title("Localisation des incidents")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.kdeplot(
    data=df,
    x="Longitude",
    y="Latitude",
    fill=True,
    cmap="Reds"
)
plt.title("Densité spatiale des crimes (KDE)")
plt.show()
```
